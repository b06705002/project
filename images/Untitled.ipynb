```python
import pandas as pd
import numpy as np
to_count = "./weights.csv"
```

```python
df = pd.read_csv(to_count)
```

```python
print(df)
```

```python
cm_list = []
c = 0
for s in df['heights']:
    if s != '0':
#         print(c)
#         print(s)
        stri = s[0:1]
        x = s.find('"')
        xtri = s[3:x]
    #     print(x)
    #     print(stri)
        st = float(stri)
    #     print(xtri)
        xt = float(xtri)
        cm = (st*12 + xt) * 2.54
#         print(cm)
        cm_list.append(cm)
        c += 1
    else:
        cm_list.append(-1)
        c += 1
cml = np.array(cm_list)
print(cml)
```

```python
w = df['weights(kg)']
BMI_list = []
for i in range(0, len(cml)):
#     print(w[i])
    bmi = float(w[i]) / ((cml[i] / 100)**2)
    BMI_list.append(bmi)
bml = np.array(BMI_list)
print(bml)
```

```python
dfc = pd.DataFrame({'height': cml/100,
                    'weight': w,
                   'BMI': bml})
```
